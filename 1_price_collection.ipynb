```python
def parse():
    try:
        RTMS_ID = row.find("RTMS_ID").get_text()
        LAND_CD = row.find("LAND_CD").get_text()
        SGG_NM = row.find("SGG_NM").get_text()
        ACC_YEAR = row.find("ACC_YEAR").get_text()
        DEAL_YMD = row.find("DEAL_YMD").get_text()
        BLDG_AREA = row.find("BLDG_AREA").get_text()
        FLR_INFO = row.find("FLR_INFO").get_text()
        BLDG_MUSE_NM = row.find("BLDG_MUSE_NM").get_text()
        OBJ_AMT = row.find("OBJ_AMT").get_text()
        BLDG_NM = row.find("BLDG_NM").get_text()
        BUILD_YEAR = row.find("BUILD_YEAR").get_text()

        return {
            "실거래가아이디":RTMS_ID,
            "지번코드":LAND_CD,
            "자치구명":SGG_NM,
            "신고년도":ACC_YEAR,
            "계약일자":DEAL_YMD,
            "건물면적":BLDG_AREA,          
            "층":FLR_INFO,    
            "건물주용도":BLDG_MUSE_NM,
            "물건금액":OBJ_AMT,
            "건물명":BLDG_NM,
            "건축년도":BUILD_YEAR
        }
    except AttributeError as e:
        return {
            "실거래가아이디":None,
            "지번코드":None,
            "자치구명":None,
            "신고년도":None,
            "계약일자":None,
            "건물면적":None,
            "층":None,
            "건물주용도":None,
            "물건금액":None,
            "건물명":None,
            "건축년도":None
        }
```

```python
price = []
# 읽어 들인 csv파일 내용을 저장할 빈 리스트를 하나 만든다

#csv 참고
#https://rfriend.tistory.com/250
for file in allFile_list:
    df = pd.read_csv(file, header=0) # for구문으로 csv파일들을 읽어 들인다, 헤더지정
    price.append(df) # 빈 리스트에 읽어 들인 내용을 추가한다
```

```python
all_price = pd.concat(price, axis=0, ignore_index=True)
```
