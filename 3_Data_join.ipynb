```python
#실거래가
price=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apt/old/apt_price_API_20062021_0118.csv', encoding='utf-8')
```

```python
#건축물 에너지 효율 등급
beec=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apt/old/beec_API_1228.csv', encoding='utf-8')
```

```python
#실거래가 beec 병합
price_beec=pd.merge(price, beec, left_on='지번코드2', right_on='지번코드2', how='left')
```

```python
price_beec = price_beec.drop_duplicates('실거래가아이디', keep='last')
```

```python
price_beec
```

```python
price_beec.columns
```

```python
price_beec = price_beec[['지번코드층연도','실거래가아이디', '신고년도', '계약일자', '자치구명', '지번코드', '건물명_x', '건물명_y', '대지위치','물건금액', '건물면적','시군구코드', '용적률', '건폐율', '총주차수', '에너지효율등급', '에너지절감률', '친환경건축물등급', '친환경건축물인증점수', 'EPI점수','지능형건축물등급', '지능형건축물인증점수', '건축년도', '층', '건물주용도' , '세대수', '생성일자']]
```

```python
#공시
gongsi=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apt/kukto/gongsi_API2_avg.csv', encoding='utf-8')
```

```python
#실거래가 공시 병합
price_beec_gongsi=pd.merge(price_beec, gongsi, left_on='지번코드층연도', right_on='지번코드층연도', how='left')
```

```python
price_beec_gongsi
```

```python
price_beec_gongsi = price_beec_gongsi.drop_duplicates('실거래가아이디', keep='last')
```

```python
price_beec_gongsi = price_beec_gongsi[[ '실거래가아이디','지번코드','신고년도', '계약일자', '자치구명', '건물명_x', '건물명_y', '대지위치','물건금액','건물면적','공시가격층평균','용적률', '건폐율', '총주차수', '에너지효율등급', '에너지절감률', '친환경건축물등급', '친환경건축물인증점수', 'EPI점수','지능형건축물등급', '지능형건축물인증점수', '건축년도', '층', '건물주용도' , '세대수']]
```

```python
df=price_beec_gongsi.copy()
```

```python
df['m2당가격']=round(df['물건금액']/df['건물면적'],3)
```

```python
df['주차스트레스']=round(df['총주차수']/df['세대수'],3)
```

```python
df.주차스트레스
```

```python
df = df[[ '지번코드','신고년도', '계약일자', '자치구명', '건물명_x', '대지위치','물건금액','m2당가격', '공시가격층평균','건물면적','용적률', '건폐율', '총주차수','주차스트레스', '에너지효율등급', '에너지절감률', '친환경건축물등급', '친환경건축물인증점수', 'EPI점수', '건축년도', '층', '세대수']]
```
